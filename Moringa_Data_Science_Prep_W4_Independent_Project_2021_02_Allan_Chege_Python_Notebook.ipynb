```
# We will first load the sql extension into our environment
%load_ext sql

# Then connect to our in memory sqlite database
# NB: This database will cease to exist as soon as the database connection is closed. 
%sql sqlite://
```

```
# Importing Pandas
import pandas as pd
# Importing numpy
import numpy as np
```

```
# Loading the table from a local CSV file
#
with open('Autolib_dataset (2).csv','r') as f:
    auto_dataset = pd.read_csv(f, index_col=0, encoding='utf-8') 

# Then storing it in an SQL table of our in memory sqlite database
#
%sql DROP TABLE if EXISTS auto_dataset;
%sql PERSIST auto_dataset;
```

```
#Format the date to an apporpriate hrs/mins format 
#start by creating a column with the new date/time format
#
%%sql
ALTER TABLE auto_dataset
ADD "Time_updated";
```

```
# update the time in the table to the right format  
%%sql
UPDATE auto_dataset
set "Time_updated" = substr("hour", 0, 3) 
|| ":" || substr("minute", 0, 3 ) 
```

```
#sort the table to find the most poplar time of day when bluecars have been picked when bluecars have been picked
%%sql
SELECT * FROM auto_dataset
WHERE month = 4 and year = 2018 
ORDER BY "Bluecar counter" DESC limit 50;
```
